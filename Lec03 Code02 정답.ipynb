## Problem 3. Merge & Join

```python
# 주문정보 거래데이터 생성하기
df_order = pd.DataFrame(
    {
        'CUSTOMER_ID': ['100', '100', '200', '200', '300'],
        'ORDER_ID': ['11', '12', '13', '14', '15'],
        'PRODUCT_CODE': ['9001', '9002', '9001', '9003', '9003']
    }
)
df_order
```

```python
# 고객정보 데이터 생성하기
df_customer = pd.DataFrame(
    {
        'CUSTOMER_ID': ['100', '200', '300', '400'],
        'NAME': ['kim', 'park', 'lee', 'choi']
    }
)
df_customer
```

```python
# 상품정보 데이터 생성하기
df_product = pd.DataFrame(
    {
        'PRODUCT_CODE': ['9001', '9003'],
        'NAME': ['Notebook', 'TV']
    }
)
df_product
```

### 1) 주문정보 테이블인 df_order와 고객정보 테이블인 df_customer를 customer_id 기반으로 merge하여 df_merge1 생성하기
### ---> 그리고 df_merge1 출력

```python
# 3.1) 
# df_order와 df_customer를 customer_id 기준준으로 merge하기
df_merge1 = pd.merge(df_order, df_customer, on='CUSTOMER_ID')

# df_merge1 출력
df_merge1
```

### 2) 주문정보 테이블인 df_order와 상품정보 테이블인 df_product를 product_id 기반으로 merge하여 df_merge2 생성하기
### (단, merge의 join 방식은 left join을 활용할 것!) 
### ---> 그리고 df_merge2 출력

```python
# 3.2) 
# df_order와 df_product를 product_code 기준으로 merge하기 ( 단, join 방식은 left join을 활용! )
df_merge2 = pd.merge(df_order, df_product, on='PRODUCT_CODE', how='left')

# df_merge2 출력
df_merge2
```

### 3) df_merge1과 df_merge2를 order_id 기반으로 join하여 df_merge3 생성하기
### ( 주의할점 1: order_id를 set_index를 활용하여 index변수로 설정하여 새로운 df_merge1_index, df_merge2_index 데이터프레임 생성하기 )
### ( 주의할점 2: merge의 join 방식은 outer join을 활용할 것!) 
### ( 주의할점 3: 공통 변수들은 suffix를 각각 customer와 product로 주기 )
### ---> 그리고 df_merge3 출력

```python
# 3.3) 
# df_merge1와 df_merge2를 order_id 기준으로 merge하기
# set_index를 활용하여 'ORDER_ID' 변수 index로 설정하여 새로운 df_merge1_index, df_merge2_index 생성하기
df_merge1_index = df_merge1.set_index('ORDER_ID')
df_merge2_index = df_merge2.set_index('ORDER_ID')
# join 진행시, 
# outer join 활용할 것
# left suffix는 '_CUSTOMER', right suffix는 '_PRODUCT'를 붙여줄 것!
df_merge3 = df_merge1_index.join(df_merge2_index, how='outer', lsuffix='_CUSTOMER', rsuffix='_PRODUCT')

# df_merge3 출력
df_merge3
```

```python
# 네이버 금융 삼성전자 2021년 주가 크롤링
import pandas_datareader.data as web
import datetime
start=datetime.datetime(2021,1,1)
end=datetime.datetime(2021,12,31)
df_samsung = web.DataReader('005930','naver',start,end)
df_samsung.head()
```

1) 2021년 3월 23일부터 2021년 4월 26일까지 데이터 슬라이싱해서 df_samsung1 생성하기

```python
df_samsung1 = df_samsung['2021-03-23':'2021-04-26']
df_samsung1.head()
```

2) 위 기간동안의 종가(Close Price) 기반 ROI 계산해보기
=> 실습 소스코드 참고

```python
# ROI는 투자대비 수익률(%) = 100 * (마지막 가격/시작 가격 - 1)
ROI = 100 * (
    int(df_samsung1['Close'][-1]) / int(df_samsung1['Close'][0]) - 1
)
# 약 1달간, 2.08% 수익률이 발생
ROI
```
