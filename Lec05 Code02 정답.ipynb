## 데이터 불러오기

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df
```

##  1) 품종별로 그룹화 한 후 히스토그램 그리기

```python
# 그룹화
df1 = df.groupby('variety')
# 히스토그램 그리기
df1.hist()
```

## 2) 품종별로 그룹화하여 평균 , 중간값, 표준편차, 백분위 구하기

```python
print('------------------------------mean-------------------------------')
print(df1.mean())
print('-----------------------------median------------------------------')
print(df1.median())
print('------------------------------std-------------------------------')
print(df1.std())
print('----------------------------quantile-----------------------------')
print(df1.quantile())
```

## 3) sepal length, sepal width, petal length, petal width 별로 mean, trim mean, weighted mean, median 값 확인하기


```python
from scipy import stats
import numpy as np

print("sepal.width")
print("mean:",df['sepal.width'].mean())
a=stats.trim_mean(df['sepal.width'],0.1)
print("trim mean:", a)
w = [i for i in range(0,150)]
b=np.average(df['sepal.width'],weights=w)
print("weighted mean:", b)
print("median:",df['sepal.width'].median())
print("\n")
print("sepal.length")
print("mean:",df['sepal.length'].mean())
a1=stats.trim_mean(df['sepal.length'],0.1)
print("trim mean:", a1)
w1 = [i for i in range(0,150)]
b1=np.average(df['sepal.length'],weights=w1)
print("weighted mean:", b1)
print("median:",df['sepal.length'].median())
print("\n")
print("petal.width")
print("mean:",df['petal.width'].mean())
a2=stats.trim_mean(df['petal.width'],0.1)
print("trim mean:", a2)
w2 = [i for i in range(0,150)]
b2=np.average(df['petal.width'],weights=w2)
print("weighted mean:", b2)
print("median:",df['petal.width'].median())
print("\n")
print("petal.length")
print("mean:",df['petal.length'].mean())
a3=stats.trim_mean(df['petal.length'],0.1)
print("trim mean:", a3)
w3 = [i for i in range(0,150)]
b3=np.average(df['petal.length'],weights=w3)
print("weighted mean:", b3)
print("median:",df['petal.length'].median())
```

## 4-1) sepal length, sepal width, petal length, petal width 별로 value counts 구하기

```python
print("sepal length")
print(df['sepal.length'].value_counts())
print('sepal width')
print(df['sepal.width'].value_counts())
print('petal length')
print(df['petal.length'].value_counts())
print('petal width')
print(df['petal.width'].value_counts())
```

## 4-2) sepal length, sepal width, petal length, petal width 별로 value counts 막대 그래프로 그리기

```python
import matplotlib.pyplot as plt

f, axes = plt.subplots(4,1)
f.set_size_inches((10, 40))
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)

ax1=df['sepal.width'].value_counts().plot(kind='barh',ax=axes[0],color='orange')
ax1.set_title('sepal width')
ax1.set_xlabel('sepal width')
ax1.set_ylabel('value counts')

ax2=df['sepal.length'].value_counts().plot(kind='barh',ax=axes[1])
ax2.set_title('sepal length')
ax2.set_xlabel('sepal length')
ax2.set_ylabel('value counts')

ax3=df['petal.width'].value_counts().plot(kind='barh',ax=axes[2],color='red')
ax3.set_title('petal width')
ax3.set_xlabel('petal width')
ax3.set_ylabel('value counts')

ax4=df['petal.length'].value_counts().plot(kind='barh',ax=axes[3],color='purple')
ax4.set_title('petal length')
ax4.set_xlabel('petal length')
ax4.set_ylabel('value counts')
```

## 5) sepal length, sepal width, petal length, petal width 별 IQR (Inter Quartile Range) 계산하기

```python
percentile75_sl=df['sepal.length'].quantile(0.75)
percentile25_sl=df['sepal.length'].quantile(0.25)
IQR_sl = percentile75_sl - percentile25_sl
print('IQR of sepal length:',IQR_sl)

percentile75_sw=df['sepal.width'].quantile(0.75)
percentile25_sw=df['sepal.width'].quantile(0.25)
IQR_sw = percentile75_sw - percentile25_sw
print('IQR of sepal width:',IQR_sw)

percentile75_pl=df['petal.length'].quantile(0.75)
percentile25_pl=df['petal.length'].quantile(0.25)
IQR_pl = percentile75_pl - percentile25_pl
print('IQR of petal length:',IQR_pl)

percentile75_pw=df['petal.width'].quantile(0.75)
percentile25_pw=df['sepal.length'].quantile(0.25)
IQR_pw = percentile75_pw - percentile25_pw
print('IQR of petal width:',IQR_pw)
```
